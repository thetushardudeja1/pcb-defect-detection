## Dataset Overview (DeepPCB)

The DeepPCB dataset is a publicly available benchmark dataset designed for
printed circuit board (PCB) defect analysis. It contains high-resolution PCB
images along with defect annotations that indicate the location and type of
manufacturing defects.

Each PCB image is accompanied by an annotation file containing bounding box
coordinates and a defect class label. A single PCB image may contain multiple
defects, and each defect is treated as an independent sample by cropping the
corresponding region using the provided bounding box.

In this project, the dataset is converted into a patch-based classification
problem. Defect regions are cropped from the original PCB images using bounding
boxes, resized to a fixed resolution, and used as inputs to a convolutional
neural network. This approach allows the model to focus on local defect patterns
instead of processing the entire PCB image.

The dataset includes six classes:
Open, Short, Crack, Missing Hole, Spur, and Background. Class labels are mapped
to integer indices for model training.

To ensure consistency between data inspection and training, the same dataset
loader and preprocessing pipeline are reused across notebooks. Data augmentation
is applied only during training to improve robustness to orientation and
appearance variations, while validation data remains unaltered to provide a
fair evaluation of model generalization.
